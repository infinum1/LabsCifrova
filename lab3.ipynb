```python
import cv2
import matplotlib.pyplot as plt
```

```python

img_half = cv2.imread('Lab3_half100.png', cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread('Lab3_chess100.png', cv2.IMREAD_GRAYSCALE)
```

```python
#img_half
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
#img_chess
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_half, cmap='gray'); 
plt.subplot(122)

plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_chess, cmap='gray'); 
plt.subplot(122)

plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); plt.show();
```

```python
half_blur3 = cv2.blur(img_half, (3,3))
half_blur5 = cv2.blur(img_half, (5,5))
half_blur9 = cv2.blur(img_half, (9,9))
half_blur15 = cv2.blur(img_half, (15,15))

chess_blur3 = cv2.blur(img_chess, (3,3))
chess_blur5 = cv2.blur(img_chess, (5,5))
chess_blur9 = cv2.blur(img_chess, (9,9))
chess_blur15 = cv2.blur(img_chess, (15,15))
```

```python
half_blur3_hist = cv2.calcHist([half_blur3], [0], None, [256], [0, 256])
half_blur5_hist = cv2.calcHist([half_blur5], [0], None, [256], [0, 256])
half_blur9_hist = cv2.calcHist([half_blur9], [0], None, [256], [0, 256])
half_blur15_hist = cv2.calcHist([half_blur15], [0], None, [256], [0, 256])

chess_blur3_hist = cv2.calcHist([chess_blur3], [0], None, [256], [0, 256])
chess_blur5_hist = cv2.calcHist([chess_blur5], [0], None, [256], [0, 256])
chess_blur9_hist = cv2.calcHist([chess_blur9], [0], None, [256], [0, 256])
chess_blur15_hist = cv2.calcHist([chess_blur15], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(20,20))
plt.subplot(321),plt.imshow(half_blur3, cmap='gray'),plt.title('Blur 3x3')
plt.xticks([]), plt.yticks([])
plt.subplot(322)
plt.plot(half_blur3_hist); plt.xlim([0, 255]); plt.grid(); plt.show();
plt.subplot(323),plt.imshow(half_blur5, cmap='gray'),plt.title('Blur 5x5')
plt.xticks([]), plt.yticks([])
plt.subplot(324)
plt.plot(half_blur5_hist); plt.xlim([0, 255]); plt.grid(); plt.show();
plt.subplot(325),plt.imshow(half_blur9, cmap='gray'),plt.title('Blur 9x9')
plt.xticks([]), plt.yticks([])
plt.subplot(326)
plt.plot(half_blur9_hist); plt.xlim([0, 255]); plt.grid(); plt.show();
plt.subplot(321),plt.imshow(half_blur15, cmap='gray'),plt.title('Blur 15x15')
plt.xticks([]), plt.yticks([])
plt.subplot(322)
plt.plot(half_blur15_hist); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(20,20))
plt.subplot(321),plt.imshow(chess_blur3, cmap='gray'),plt.title('Blur 3x3')
plt.xticks([]), plt.yticks([])
plt.subplot(322)
plt.plot(chess_blur3_hist); plt.xlim([0, 255]); plt.grid(); plt.show();
plt.subplot(321),plt.imshow(chess_blur5, cmap='gray'),plt.title('Blur 5x5')
plt.xticks([]), plt.yticks([])
plt.subplot(322)
plt.plot(chess_blur5_hist); plt.xlim([0, 255]); plt.grid(); plt.show();
plt.subplot(321),plt.imshow(chess_blur9, cmap='gray'),plt.title('Blur 9x9')
plt.xticks([]), plt.yticks([])
plt.subplot(322)
plt.plot(chess_blur9_hist); plt.xlim([0, 255]); plt.grid(); plt.show();
plt.subplot(321),plt.imshow(chess_blur15, cmap='gray'),plt.title('Blur 15x15')
plt.xticks([]), plt.yticks([])
plt.subplot(322)
plt.plot(chess_blur15_hist); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.show()

```
